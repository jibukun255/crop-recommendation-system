```python
"""
00_topo_extraction.py — Step 1: Topography Feature Extraction
==============================================================
Reads the raw dataset, queries NASADEM (via Open-Elevation API)
for each row's lat/lon, then computes and appends:
  - elevation  (metres above sea level)
  - slope      (terrain steepness in degrees)
  - aspect     (compass direction the slope faces, in degrees)

Output: data/raw/crop_data_with_topo.csv

Usage:
    python src/00_topo_extraction.py
"""

import os
import sys
import time
import math
import requests
import pandas as pd

# ── Path setup ────────────────────────────────────────────────────────────────
BASE_DIR   = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
INPUT_CSV  = os.path.join(BASE_DIR, "data", "raw", "crop_data.csv")
OUTPUT_CSV = os.path.join(BASE_DIR, "data", "raw", "crop_data_with_topo.csv")

# ── API settings ──────────────────────────────────────────────────────────────
API_URL    = "https://api.open-elevation.com/api/v1/lookup"
BATCH_SIZE = 100      # rows per API request (API limit is ~500, keep low to be safe)
RETRY_MAX  = 3        # retries on failure
RETRY_WAIT = 5        # seconds between retries
NODATA_VAL = -32767   # NASADEM sentinel for missing terrain data

# ── Column names in the raw dataset ──────────────────────────────────────────
LAT_COL = "latitude"
LON_COL = "longitude"


# ─────────────────────────────────────────────────────────────────────────────
# 1. Elevation fetching
# ─────────────────────────────────────────────────────────────────────────────

def fetch_elevations_batch(lat_lon_pairs: list[tuple]) -> list[float]:
    """
    Send one batch of (lat, lon) pairs to Open-Elevation API.
    Returns a list of elevation values (metres).
    Falls back to NODATA_VAL if the request fails after retries.
    """
    locations = [{"latitude": lat, "longitude": lon} for lat, lon in lat_lon_pairs]
    payload   = {"locations": locations}

    for attempt in range(1, RETRY_MAX + 1):
        try:
            response = requests.post(API_URL, json=payload, timeout=30)
            response.raise_for_status()
            results = response.json().get("results", [])
            return [r.get("elevation", NODATA_VAL) for r in results]
        except Exception as e:
            print(f"    Attempt {attempt}/{RETRY_MAX} failed: {e}")
            if attempt < RETRY_MAX:
                print(f"    Retrying in {RETRY_WAIT}s...")
                time.sleep(RETRY_WAIT)

    # All retries failed — return NODATA for every row in this batch
    print(f"    Batch failed permanently. Filling {len(lat_lon_pairs)} rows with NODATA.")
    return [NODATA_VAL] * len(lat_lon_pairs)


def fetch_all_elevations(df: pd.DataFrame) -> list[float]:
    """
    Fetch elevations for the entire dataframe in batches.
    Prints progress as it goes.
    """
    all_elevations = []
    total          = len(df)
    pairs          = list(zip(df[LAT_COL], df[LON_COL]))

    print(f"\n  Fetching elevations for {total:,} rows in batches of {BATCH_SIZE}...")

    for start in range(0, total, BATCH_SIZE):
        end   = min(start + BATCH_SIZE, total)
        batch = pairs[start:end]

        print(f"  Rows {start+1:>4} – {end:>4} / {total}", end="  ")
        elevations = fetch_elevations_batch(batch)
        all_elevations.extend(elevations)
        print(f"done  (sample elevation: {elevations[0]:.0f} m)")

        # Be polite to the API — short pause between batches
        if end < total:
            time.sleep(0.5)

    return all_elevations


# ─────────────────────────────────────────────────────────────────────────────
# 2. Slope & Aspect computation
# ─────────────────────────────────────────────────────────────────────────────

def compute_slope_aspect(df: pd.DataFrame) -> tuple[list[float], list[float]]:
    """
    Approximates slope and aspect for each row using its nearest neighbours
    in the dataset (based on geographic proximity).

    Method:
      For each point P, find its closest neighbour N.
      - Slope   = arctan(|elev_P - elev_N| / distance_PN)  in degrees
      - Aspect  = compass bearing from P to N               in degrees (0=N, 90=E)

    This is a simplified approximation suitable for a student project.
    A production system would use a proper DEM raster grid.
    """
    from sklearn.neighbors import BallTree

    lats  = df[LAT_COL].values
    lons  = df[LON_COL].values
    elevs = df["elevation"].values

    # Build a BallTree on lat/lon in radians for fast nearest-neighbour lookup
    coords_rad = [[math.radians(la), math.radians(lo)] for la, lo in zip(lats, lons)]
    tree       = BallTree(coords_rad, metric="haversine")

    # Query 2 neighbours: [0] is the point itself, [1] is the closest other point
    distances, indices = tree.query(coords_rad, k=2)

    R_EARTH = 6371000  # metres

    slopes  = []
    aspects = []

    for i in range(len(df)):
        neighbour_idx  = indices[i][1]
        dist_rad       = distances[i][1]
        dist_m         = dist_rad * R_EARTH  # convert radians → metres

        elev_diff = abs(float(elevs[i]) - float(elevs[neighbour_idx]))

        # Slope: angle of elevation change over horizontal distance
        if dist_m > 0:
            slope_deg = math.degrees(math.atan(elev_diff / dist_m))
        else:
            slope_deg = 0.0

        # Aspect: compass bearing from point i to its nearest neighbour
        lat1 = math.radians(lats[i]);           lon1 = math.radians(lons[i])
        lat2 = math.radians(lats[neighbour_idx]); lon2 = math.radians(lons[neighbour_idx])
        dlon = lon2 - lon1
        x    = math.sin(dlon) * math.cos(lat2)
        y    = math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2) * math.cos(dlon)
        bearing = (math.degrees(math.atan2(x, y)) + 360) % 360

        slopes.append(round(slope_deg, 6))
        aspects.append(round(bearing, 6))

    return slopes, aspects


# ─────────────────────────────────────────────────────────────────────────────
# 3. Missing value handling
# ─────────────────────────────────────────────────────────────────────────────

def handle_nodata(df: pd.DataFrame) -> pd.DataFrame:
    """
    Replaces NASADEM NoData values (-32767) using nearest-neighbour
    interpolation within the same crop class. Rows that cannot be
    interpolated are dropped.
    """
    nodata_mask = df["elevation"] == NODATA_VAL
    nodata_count = nodata_mask.sum()

    if nodata_count == 0:
        print("  No NoData values found — all elevations valid.")
        return df

    print(f"  Found {nodata_count} NoData elevation(s). Attempting interpolation...")

    for idx in df[nodata_mask].index:
        crop  = df.at[idx, "label"]
        # Use mean elevation of same-crop records that have valid data
        valid = df[(df["label"] == crop) & (df["elevation"] != NODATA_VAL)]
        if not valid.empty:
            df.at[idx, "elevation"] = int(valid["elevation"].mean())
        # else: leave as NODATA, will be dropped below

    # Drop any rows still carrying NODATA (couldn't interpolate)
    remaining = (df["elevation"] == NODATA_VAL).sum()
    if remaining > 0:
        print(f"  Dropping {remaining} row(s) that could not be interpolated.")
        df = df[df["elevation"] != NODATA_VAL].reset_index(drop=True)

    print(f"  NoData handling complete. Dataset now has {len(df):,} rows.")
    return df


# ─────────────────────────────────────────────────────────────────────────────
# 4. Main
# ─────────────────────────────────────────────────────────────────────────────

def main():
    print("=" * 62)
    print("  Step 0 — Topography Feature Extraction")
    print("=" * 62)

    # ── Load raw dataset ──────────────────────────────────────────
    if not os.path.exists(INPUT_CSV):
        print(f"\n  ERROR: Raw dataset not found at:\n    {INPUT_CSV}")
        print("  Download from: https://data.mendeley.com/datasets/8v757rr4st/1")
        sys.exit(1)

    df = pd.read_csv(INPUT_CSV)
    print(f"\n  Loaded: {INPUT_CSV}")
    print(f"  Shape : {df.shape[0]:,} rows × {df.shape[1]} columns")

    # ── Validate required columns ─────────────────────────────────
    missing_cols = [c for c in [LAT_COL, LON_COL] if c not in df.columns]
    if missing_cols:
        print(f"\n  ERROR: Missing columns: {missing_cols}")
        print(f"  Available columns: {list(df.columns)}")
        sys.exit(1)

    # ── Check if already extracted ────────────────────────────────
    if all(c in df.columns for c in ["elevation", "slope", "aspect"]):
        print("\n  Topography columns already present — skipping extraction.")
        print(f"  Saving as-is to: {OUTPUT_CSV}")
        df.to_csv(OUTPUT_CSV, index=False)
        return

    # ── Step 1: Fetch elevations ──────────────────────────────────
    df["elevation"] = fetch_all_elevations(df)

    # ── Step 2: Handle NASADEM NoData ─────────────────────────────
    print("\n── NoData treatment ───────────────────────────────────────")
    df = handle_nodata(df)

    # ── Step 3: Compute slope & aspect ───────────────────────────
    print("\n── Computing slope and aspect ─────────────────────────────")
    print("  Using nearest-neighbour approximation...")
    slopes, aspects = compute_slope_aspect(df)
    df["slope"]  = slopes
    df["aspect"] = aspects
    print("  Done.")

    # ── Step 4: Save output ───────────────────────────────────────
    df.to_csv(OUTPUT_CSV, index=False)

    print("\n── Summary ────────────────────────────────────────────────")
    print(f"  Output saved : {OUTPUT_CSV}")
    print(f"  Final shape  : {df.shape[0]:,} rows × {df.shape[1]} columns")
    print(f"  New columns  : elevation, slope, aspect")
    print(f"\n  Elevation stats:")
    print(f"    Min  : {df['elevation'].min():>8.0f} m")
    print(f"    Max  : {df['elevation'].max():>8.0f} m")
    print(f"    Mean : {df['elevation'].mean():>8.0f} m")
    print(f"\n  Slope stats:")
    print(f"    Min  : {df['slope'].min():>8.4f} °")
    print(f"    Max  : {df['slope'].max():>8.4f} °")
    print(f"    Mean : {df['slope'].mean():>8.4f} °")
    print("\n" + "=" * 62)
    print("  Next step: python src/01_preprocess.py")
    print("=" * 62)


if __name__ == "__main__":
    main()
```
